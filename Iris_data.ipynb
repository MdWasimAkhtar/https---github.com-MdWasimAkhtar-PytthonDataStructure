bar plot:  Here the frequency of the observation is plotted.In this case we are plotting the frequency of the three species in the Iris Dataset

```python
sea.countplot(data=iris_data)
```

```python
sea.countplot('Species', data=iris_data)
```

pie Chart :

```python
iris_data['Species'].value_counts().plot.pie(explode=[0.1,0.1,0.1],autopct="%1.1f%%",shadow=True , figsize=(15,8))
```

We can see that there are 50 samples each of all the Iris Species in the data set.

join plot: Jointplot is seaborn library specific and can be used to quickly visualize and analyze the relationship between two variables and describe their individual distributions on the same plot.

```python
sea.jointplot(x='SepalLengthCm', y="SepalWidthCm",data=iris_data)
```

```python
sea.pairplot(iris_data , kind="reg")
```

```python
sea.pairplot(iris_data , hue="Species" , corner=True)
```

Assigning a hue variable adds a semantic mapping and changes the default marginal plot to a layered kernel density estimate (KDE):

kind{‘scatter’, ‘kde’, ‘hist’, ‘reg’}

```python
sea.jointplot("SepalLengthCm","SepalWidthCm",data=iris_data,kind="kde")
```

. Boxplot or Whisker plot Box plot was was first introduced in year 1969 by Mathematician John Tukey.Box plot give a statical summary of the features being plotted.Top line represent the max value,top edge of box is third Quartile, middle edge represents the median,bottom edge represents the first quartile value.The bottom most line respresent the minimum value of the feature.The height of the box is called as Interquartile range.The black dots on the plot represent the outlier values in the data.

```python
fig = plt.gcf()
fig.set_size_inches(10,7)
sea.boxplot(x="Species" , y="PetalLengthCm" , data=iris_data , linewidth=2   )
```

```python
iris_data.boxplot(by='Species' , figsize=(12,6))
```
