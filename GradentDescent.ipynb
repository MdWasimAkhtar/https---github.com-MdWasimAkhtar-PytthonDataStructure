# A simple Cost Function 
##    $f(x) = X^2 + x + 1$

```python
def f(x):
    return x**2 + x + 1;
```

```python
# generate data
X_1 = np.linspace(start=-3 , stop=3 , num=500)
#X_1[2]
```

```python
#plot the data 

plt.xlim(-6,6)
plt.ylim(-2,8)
plt.plot(X_1 , f(X_1) , color='red')

plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)',fontsize=20)
plt.show()
```

## Slop and Derivative 
  slop of the function is find out by derivative
 create a python function for derivative of f(x) called d(x)

```python
def df(x):
    return  2*x + 1 
```

```python
plt.figure(figsize=[10,10])

#plt.subplot(1, 2, 1)
#the figure has 1 row, 2 columns, and this plot is the first plot.

plt.subplot(1,2,1)

plt.plot(X_1 , f(X_1) , color='red' , linewidth=4)

plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)',fontsize=20)
#plt.show()

# 2nd plot
plt.subplot(1, 2, 2)


plt.plot(X_1 , df(X_1)  , color="blue" , linewidth=3)
plt.grid()
plt.xlabel('X' , fontsize=18)
plt.ylabel('df(x)' , fontsize=20)

plt.show()
```

```python

plt.plot(X_1 , df(X_1)  , color="blue" , linewidth=3)
plt.grid()
plt.xlabel('X' , fontsize=18)
plt.ylabel('df(x)' , fontsize=20)
plt.xlim(-6,6)
plt.show()
```

## python Loop & Gradient Descent

```python
# gradient descent
new_x= -3 
previous_x=0
step_multipiler = 0.1
precision = .00001

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(800 ):
    previous_x = new_x
    gradient = df(previous_x)
    new_x = previous_x - step_multipiler * gradient

    step_size= abs(new_x - previous_x)

    x_list.append(new_x)
    slope_list.append(df(new_x))

    #print('step_size : ',step_size)
    if(step_size < precision):
        print('loop run :',n)
        break
       



print('local_minimun occurs at :', new_x)
print('slope of df(x) value at this pont is : ',df(new_x))
print('f(x) value or cost at this point is : ',f(new_x))
```

```python
#super-impose the gradient Descent

plt.figure(figsize=[20,10])

#plt.subplot(1, 2, 1)
#the figure has 1 row, 2 columns, and this plot is the first plot.

plt.subplot(1,2,1)

plt.plot(X_1 , f(X_1) , color='blue' , linewidth=4)

plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)',fontsize=20)

values = np.array(x_list)
plt.scatter(x_list , f(values) , color="red" , s= 100 , alpha=0.4)

# 2nd plot
plt.subplot(1, 2, 2)


plt.plot(X_1 , df(X_1)  , color="blue" , linewidth=3)
plt.grid()
plt.xlabel('X' , fontsize=18)
plt.ylabel('df(x)' , fontsize=20)
plt.xlim(-6,6)
plt.scatter(x_list , slope_list, color='red' , s=100 )
plt.show()
```

## example2- multiple Minima vs Initial Guess and Adv finction
## $$g(x) = x^4 - 4*x^2 + 5$$

```python
# make some data
x_2 = np.linspace(-2 , 2 , 1000)

#
def g(x):
    return x**4 - 4 * x**2 + 5 

def dg(x):
    return 4*x**3 -8*x
```

```python
plt.figure(figsize=[20,10])

#plt.subplot(1, 2, 1)
#the figure has 1 row, 2 columns, and this plot is the first plot.

plt.subplot(1,2,1)


plt.plot(x_2 , g(x_2) , color='red' , linewidth=4)

plt.xlabel('X', fontsize=16)
plt.ylabel('g(x)',fontsize=20)
#plt.show()

# 2nd plot
plt.subplot(1, 2, 2)


plt.plot(x_2 , dg(x_2)  , color="blue" , linewidth=3)
plt.grid()
plt.xlabel('X' , fontsize=18)
plt.ylabel('df(x)' , fontsize=20)

plt.show()
```

```python
def gradient_descent(derivative_func, initial_guess, multiplier , precision):


# gradient descent
 
    new_x= initial_guess
   

    x_list = [new_x]
    slope_list = [df(new_x)]

    for n in range(800 ):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size= abs(new_x - previous_x)

        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        #print('step_size : ',step_size)
        if(step_size < precision):
            print('loop run :',n)
            break
    return new_x , x_list , slope_list
        



print('local_minimun occurs at :', new_x)
print('slope of df(x) value at this pont is : ',df(new_x))
print('f(x) value or cost at this point is : ',f(new_x))
```

```python
local_min , list_x , deriv_list =  gradient_descent(dg , 0.5 , 0.02 ,precision= 0.001  )
print('local min occurs at : ',local_min)
print('number of step : ', len(list_x))
```

```python
plt.figure(figsize=[20,10])

#plt.subplot(1, 2, 1)
#the figure has 1 row, 2 columns, and this plot is the first plot.

plt.subplot(1,2,1)


plt.plot(x_2 , g(x_2) , color='blue' , linewidth=4)
plt.scatter(list_x , g(np.array(list_x)),color='red', s=100 , alpha=0.6)

plt.xlabel('X', fontsize=16)
plt.ylabel('g(x)',fontsize=20)
#plt.show()

# 2nd plot
plt.subplot(1, 2, 2)


plt.plot(x_2 , dg(x_2)  , color="blue" , linewidth=3)
plt.scatter(list_x,deriv_list , color="red" , s=100 , alpha=0.6)
plt.grid()
plt.xlabel('X' , fontsize=18)
plt.ylabel('df(x)' , fontsize=20)

plt.show()
```

## example 3- Divergence , Overflow and python tuples
## $$h(x) = x^5 -2x^4 +2 $$


```python
x_3 = np.linspace(start=2.5 , stop=2.5 , num=1000)

def h(x):
    return x**5 - 2*x**4 + 2

def dh(x):
    return 5*x**4 -8*x**3 
```

```python
local_min , list_x , deriv_list =  gradient_descent(dh , 0.2 , 0.02 ,precision= 0.001  )
print('local min occurs at : ',local_min)
print('number of step : ', len(list_x))
```

```python
plt.figure(figsize=[20,10])

#plt.subplot(1, 2, 1)
#the figure has 1 row, 2 columns, and this plot is the first plot.

plt.subplot(1,2,1)

plt.xlim(-1.2, 2.5)
plt.ylim(-1 , 4)
plt.plot(x_3 , h(x_3) , color='blue' , linewidth=4)
plt.scatter(list_x , h(np.array(list_x)),color='red', s=100 , alpha=0.6)

plt.xlabel('X', fontsize=16)
plt.ylabel('h(x)',fontsize=20)
#plt.show()

# 2nd plot
plt.subplot(1, 2, 2)

plt.xlim(-1, 2)
plt.ylim(-1 , 5)

plt.plot(x_3 , dh(x_3)  , color='blue' , linewidth=3)
plt.scatter(list_x,deriv_list , color="red" , s=100 , alpha=0.6)
plt.grid()
plt.xlabel('x_3' , fontsize=18)
plt.ylabel('dh(x)' , fontsize=20)

plt.show()
```
